```python
df = pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target'] = data.target
```

```python
df
```

```python
df['target_names'] = df.target.apply(lambda x: data.target_names[x])
```

```python
df.to_csv('breast_cancer.csv')
```

```python
new_df = df.copy()
```

```python
x = df.drop(['target_names','target'],axis=1)
y = df.target
```

```python
x.shape
```

```python
y.values
```

```python
df[df.target==1]
```

```python
sns.countplot(y)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x1 = sc.fit_transform(x)
```

```python
x1
```

```python
from sklearn.model_selection import train_test_split
```

```python
x1_train,x1_test,y_train,y_test = train_test_split(x1,y,test_size=0.2,random_state=42)
```

```python
x1_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression(random_state=51,multi_class='ovr')
reg.fit(x1_train,y_train)
pred = reg.predict(x1_test)
pred
```

```python
 from sklearn.svm import SVC
```

```python
reg1 = SVC()
reg1.fit(x1_train,y_train)
```

```python
pred1 = reg1.predict(x1_test)
pred1
```

```python
reg.score(x1_test,y_test)
```

```python
reg1.score(x1_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,pred)
```

```python
cm
```

```python
vis = pd.DataFrame(cm)
vis.columns = ['Predicted 0','Predidicted 1']
vis = vis.rename(index={0:'Actual 0',1:'Actual 1'})
```
