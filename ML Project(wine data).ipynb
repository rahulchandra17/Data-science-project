```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
from sklearn.datasets import load_wine
wine_data = load_wine()
```

```python
pd.set_option('display.max_rows',None)
```

```python
dir(wine_data)
```

```python
df_wine = pd.DataFrame(wine_data.data,columns=wine_data.feature_names)
df_wine['target'] = wine_data.target
df_wine['target_name'] = df_wine.target.apply(lambda x:wine_data.target_names[x])
```

```python
df_wine
```

```python
feature_wine = df_wine.drop(['target','target_name'],axis=1)
label_wine = df_wine.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train,feature_test,label_train,label_test = train_test_split(feature_wine,label_wine,test_size=0.1,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale.fit(feature_train)
featiure_train = scale.transform(feature_train)
```

```python
from sklearn.linear_model import LogisticRegression
reg_clf = LogisticRegression()
reg_clf.fit(feature_train,label_train)
```

```python
reg_clf.score(feature_train,label_train)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(featiure_train,label_train)
```

```python
svc.score(featiure_train,label_train)
```

```python
from sklearn.model_selection import cross_val_score 
validation1 = np.mean(cross_val_score(svc,featiure_train,label_train,cv=10))
validation2 = np.mean(cross_val_score(reg_clf,featiure_train,label_train,cv=10))
print(validation1,validation2)
```
